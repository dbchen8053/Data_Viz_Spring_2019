```python
%matplotlib inline
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# File to Load (Remember to change these)
city_data_to_load = pd.read_csv ("data/city_data.csv")
ride_data_to_load = pd.read_csv ("data/ride_data.csv")

# Read the City and Ride Data
ride_data_to_load.head()
# Combine the data into a single dataset
city_ride_data = pd.merge (ride_data_to_load, city_data_to_load, how= "left", on= "city")
# Display the data table for preview
city_ride_data.head()
```

## Bubble Plot of Ride Sharing Data

```python
# Obtain the x and y coordinates for each of the three city types
urban_data = city_ride_data.loc[city_ride_data["type"] == "Urban"]

group_urban_city = urban_data.groupby ("city")

urban_total_drivers = group_urban_city ["driver_count"].sum()

urban_total_rider = group_urban_city ["ride_id"].count()

urban_total_fare = group_urban_city ["fare"].sum()

urban_count_fare = group_urban_city ["fare"].count()

urban_ave_fare = urban_total_fare/urban_count_fare




suburban_data = city_ride_data.loc[city_ride_data["type"] == "Suburban"]

group_suburban_city = suburban_data.groupby ("city")

suburban_total_drivers = group_suburban_city ["driver_count"].sum()

suburban_total_rider = group_suburban_city ["ride_id"].count()

suburban_total_fare = group_suburban_city ["fare"].sum()

suburban_count_fare = group_suburban_city ["fare"].count()

suburban_ave_fare = suburban_total_fare/suburban_count_fare



rural_data = city_ride_data.loc[city_ride_data["type"] == "Rural"]

group_rural_city = rural_data.groupby ("city")

rural_total_drivers = group_rural_city ["driver_count"].sum()

rural_total_rider = group_rural_city ["ride_id"].count()

rural_total_fare = group_rural_city ["fare"].sum()

rural_count_fare = group_rural_city ["fare"].count()

rural_ave_fare = rural_total_fare/rural_count_fare


# Build the scatter plots for each city types
urban = plt.scatter(urban_total_rider, urban_ave_fare, marker="o", facecolors="lightcoral", edgecolors="black", 
             s=urban_total_drivers, alpha=0.50, label="urban")

suburban = plt.scatter(suburban_total_rider, suburban_ave_fare, marker="o", facecolors="lightskyblue", edgecolors="black", 
             s=suburban_total_drivers, alpha=0.50, label="suburban")

rural = plt.scatter(rural_total_rider, rural_ave_fare, marker="o", facecolors="gold", edgecolors="black", 
             s=rural_total_drivers, alpha=0.50, label="rural")



lgnd = plt.legend(fontsize="small", mode="Expanded", 
                  numpoints=1, scatterpoints=1, 
                  loc="best", title="City Types", 
                  labelspacing=0.5)
lgnd.legendHandles[0]._sizes = [30]
lgnd.legendHandles[1]._sizes = [30]
lgnd.legendHandles[2]._sizes = [30]

# Incorporate the other graph properties
plt.xlabel('Total Number of Rides (Per City)')
plt.ylabel('Average Fare ($)')
plt.title ("Pyber Ride Sharing Data (2016)")



# Create a legend

# Incorporate a text label regarding circle size

# Save Figure
plt.grid()
plt.show()

plt.savefig("Pyber Rider Sharing Data (2016)")

```

## Total Fares by City Type

```python
# Calculate Type Percents
city_ride_data_urban = city_ride_data.loc[city_ride_data["type"] == "Urban"]
urban_fare = city_ride_data_urban ["fare"].sum()

city_ride_data_suburban = city_ride_data.loc[city_ride_data["type"] == "Suburban"]
suburban_fare = city_ride_data_suburban ["fare"].sum()

city_ride_data_rural = city_ride_data.loc[city_ride_data["type"] == "Rural"]
rural_fare = city_ride_data_rural ["fare"].sum()


# Build Pie Chart

labels = ["Urban", "Suburban", "Rural"]

sizes = [urban_fare, suburban_fare, rural_fare]

colors = ["lightcoral", "lightskyblue", "gold"]

explode = (0.1, 0, 0)

plt.pie(sizes, explode=explode, labels=labels, colors=colors, autopct="%1.1f%%", shadow=True, startangle=280)

plt.title("% of Total Fares by City Type")


plt.show()


# Save Figure

plt.savefig("Total Fares by City Type")
```

```python
# Calculate Ride Percents
city_ride_data_urban = city_ride_data.loc[city_ride_data["type"] == "Urban"]
urban_ride = city_ride_data_urban ["ride_id"].count()

city_ride_data_suburban = city_ride_data.loc[city_ride_data["type"] == "Suburban"]
suburban_ride = city_ride_data_suburban ["ride_id"].count()

city_ride_data_rural = city_ride_data.loc[city_ride_data["type"] == "Rural"]
rural_ride = city_ride_data_rural ["ride_id"].count()


# Build Pie Chart

labels = ["Urban", "Suburban", "Rural"]

sizes = [urban_ride, suburban_ride, rural_ride]

colors = ["lightcoral", "lightskyblue", "gold"]

explode = (0.1, 0, 0)

plt.pie(sizes, explode=explode, labels=labels, colors=colors, autopct="%1.1f%%", shadow=True, startangle=280)

plt.title("% of Total Rides by City Type")


plt.show()



# Save Figure

plt.savefig("Total Rides by City Type")
```
